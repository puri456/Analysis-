```python

df['Sex_encoded'] = df['Sex'].map({'male': 0, 'female': 1})
heatmap_df = df[['Survived', 'Pclass', 'Age', 'Fare', 'Sex_encoded']].dropna()
sns.heatmap(heatmap_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(heatmap_df, hue='Survived')
plt.suptitle("Pairplot by Survival", y=1.02)
plt.show()
```

```python
df['Age'].hist(bins=30, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Age vs Survival")
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title("Fare vs Age by Survival")
plt.show()
```
